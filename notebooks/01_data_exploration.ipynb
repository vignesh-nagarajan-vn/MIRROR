# 01 · ChestX-ray14 Data Exploration

Inspect label distribution and a few sample films before training.
Assumes the NIH release is at `datasets/raw/chestxray14/` (see `datasets/README.md`).

```python
import sys; sys.path.append('..')
import pandas as pd
from pathlib import Path
from models.common.constants import CHESTXRAY14_LABELS

ROOT = Path('../datasets/raw/chestxray14')
meta = pd.read_csv(ROOT / 'Data_Entry_2017.csv')
meta.head()
```

## Label frequencies

ChestX-ray14 is multi-label and highly imbalanced — useful to know before choosing a loss weighting strategy.

```python
counts = {l: meta['Finding Labels'].str.contains(l).sum() for l in CHESTXRAY14_LABELS}
pd.Series(counts).sort_values(ascending=False)
```

```python
# Quick bar chart
import matplotlib.pyplot as plt
pd.Series(counts).sort_values().plot.barh(figsize=(7,5))
plt.title('ChestX-ray14 label frequency'); plt.tight_layout(); plt.show()
```

## Sample images

```python
from PIL import Image
fig, axes = plt.subplots(1, 3, figsize=(11,4))
for ax, name in zip(axes, meta['Image Index'].head(3)):
    ax.imshow(Image.open(ROOT/'images'/name), cmap='gray'); ax.axis('off')
    ax.set_title(name, fontsize=8)
plt.show()
```
